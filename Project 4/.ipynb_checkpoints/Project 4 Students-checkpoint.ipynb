## Data Description and Preprocessing

For this project, you will explore data from the [National Health and Nutrition Examination Survey](https://www.kaggle.com/cdc/national-health-and-nutrition-examination-survey?select=questionnaire.csv). NHANES is a unique study that combines survey methodology with in-person medical examinations to create a dataset with demographic information, health indicators, and health outcomes.

We start by importing the data and doing some preliminary preprocessing for you. We import some libraries that will be helpful as well. 'SEQN' is the ID number for each respondent, and 'HSD010' will be our target outcome. [HSD010](https://wwwn.cdc.gov/Nchs/Nhanes/2013-2014/HSQ_H.htm#HSD010) asks for the respondent's self reported health condition, which can range from "excellent" to "poor."

```python
# Import libraries
import pandas as pd
import numpy as np
import tensorflow as tf
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from keras.models import Sequential
from keras.layers import Dense, Dropout
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from sklearn.model_selection import train_test_split
sns.set_style("darkgrid")
from sklearn.model_selection import train_test_split
```

```python
# Load nhanes data
nhanes = pd.read_csv('nhanes.csv')
# Get the ID numbers for each observation (seqn)
seqn = nhanes['SEQN']
# Get the target, "self-reported health condition," HSD010
hsd010 = nhanes['HSD010']
# Drop SEQN from the dataframe and then apply the standard scaler
nhanes = nhanes.drop(['SEQN', 'HSD010'], axis = 1)
nhanes_scaled = pd.DataFrame(StandardScaler().fit_transform(nhanes),
                             columns = nhanes.columns)
# Add the ID and target back in
nhanes_scaled['SEQN'] = seqn
nhanes_scaled['HSD010'] = hsd010
nhanes_scaled = nhanes_scaled.set_index('SEQN')
nhanes_scaled.head()
```

### Train PCA and Discuss Number of Components

```python
# Train PCA
from sklearn.decomposition import PCA

pca = PCA(n_components=5)
principalComponents_nhanes = pca.fit_transform(nhanes_scaled)
 
pca_df = pd.DataFrame(data = principalComponents_nhanes, 
                      columns = ['principal component 1', 
                                 'principal component 2',
                                 'principal component 3',
                                 'principal component 4',
                                 'principal component 5'])


pca_df.head()
```

I chose 5 components to start with to help identify distinct patterns that may relate to the 5 outcome levels of the interest vraible. By retaining 5 components, I am attempting to preserve relevant information for predicting the outcome levels while reducing dimensionality.

### Barplot of Components

```python
# Get the explained variance ratio
explained_variance_ratio = pca.explained_variance_ratio_

# Plot a barplot of the variation explained by each component
plt.bar(range(1, len(explained_variance_ratio) + 1), explained_variance_ratio)
plt.xlabel('Principal Component')
plt.ylabel('Explained Variance Ratio')
plt.title('Explained Variance Ratio by Principal Component')
plt.show()
```

### How many components will you use to fit a supervised learning model?

```python
# Choose how many components to use for a supervised learning model
num_components_for_model = 3  # You can choose based on the plot and desired variance coverage
```

Biggest drop off is from 2 to 3. I chose to cut off at feature 3 for the suoervised model because after 3 there is minimal expllained variance ratio from one principal component to the next. ### 2D Scatterplot of the first two components

```python
# Fit PCA with the chosen number of components
pca = PCA(n_components=num_components_for_model)
nhanes_pca = pca.fit_transform(nhanes_scaled)

# 2D Scatterplot of the first two components
plt.scatter(nhanes_pca[:, 0], nhanes_pca[:, 1])
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('2D Scatterplot of the First Two Components')
plt.show()
```

PC1 and PC2 exhibit a clear correlation, suggesting a direct relationship. However, there are two discernible clusters. As both PC1 and PC2 increase, the data points show increased dispersion, indicating a more  dynamic relationship between the principal components.

### Choose a Clustering Algorithm

For clustering the nhanes data, i'll use K-Means, a partitioning algorithm that assigns data points to 'k' clusters based on their similarity, minimizing the sum of squared distances to the cluster centers.

### Cluster nhanes

```python
# set random seed
np.random.seed(10)

# Cluster the nhanes data using K-Means with the elbow method
from sklearn.cluster import KMeans
from yellowbrick.cluster import KElbowVisualizer

# Identify optimal number of clusters using the elbow method
model = KMeans(n_init=10)  # You can adjust the value based on your preference
visualizer = KElbowVisualizer(model, k=(2, 10))
visualizer.fit(nhanes_scaled)
visualizer.show()

```

### Plot your clusters on top of the BMI v. Income Poverty Ratio Plot

```python
# Choose the optimal number of clusters based on the elbow method
optimal_clusters = 4  # Adjust based on the elbow method result

# Cluster the nhanes data with the chosen number of clusters
kmeans = KMeans(n_clusters=optimal_clusters, random_state=42, n_init=10)
clusters_original = kmeans.fit_predict(nhanes_scaled)

# Plot clusters on top of BMI v. Income Poverty Ratio Plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x="INDFMPIR", y="BMXBMI", hue=clusters_original, palette="viridis", data=nhanes_scaled)
plt.xlabel('Income to Poverty Line Ratio')
plt.ylabel('Body Mass Index')
plt.title('K-Means Clustering on BMI v. Income Poverty Ratio')
plt.show()
```

Clusters 2 and 3 tends to have lower income to poverty line ratio assocuated with lower BMI. While cluster 1 tends to have a higher income to poverty line ratio associated with lower BMI. Cluster 0 seens to have no clear associated between BMI and Income to Poverty Line Ratio. 

### Retrain the clustering algorithm on PCA components and plot clusters on your 2D scatter

```python
# Retrain K-Means with PCA results
kmeans_pca = KMeans(n_clusters=optimal_clusters, random_state=42, n_init=10)
clusters_pca = kmeans_pca.fit_predict(nhanes_pca)

# Plot clusters on top of 2D PCA scatterplot
plt.figure(figsize=(8, 6))
plt.scatter(nhanes_pca[:, 0], nhanes_pca[:, 1], c=clusters_pca, cmap='viridis')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('K-Means Clustering on 2D PCA Scatterplot')
plt.show()
```

Cluster zero seems to have a direct linear relationship between PC 2 and PC 1, almost equal to 1. Cluster 1 and 2 seem to have much less spread although still relativley linear. Cluster 3 seens to have less of a distinct relationship with PC 1 and PC 2 as PC1 remains mostly arond 0 or lower while there is more spread as PC 2 increases.

### Neural Network on Original Features

```python
# partition data
# -----------
mapping = {'good': 1, 'poor': 0}

# Apply the mapping to the 'hsd010_binary' column
hsd010_binary = hsd010_binary.map(mapping)
y = hsd010_binary # either hsd010 or hsd010_binary, may need to convert to numeric if it isn't already 
X = nhanes_scaled # drop out any columns that aren't features

X_train, X_test, y_train, y_test = train_test_split(X, 
                                                    y, 
                                                    test_size = .25, 
                                                    random_state = 10)
```

```python
# load libraries
# -----------
import keras
from keras.utils import to_categorical

num_classes = 2

# convert class vectors to binary class matrices
y_train = keras.utils.to_categorical(y_train, num_classes=num_classes)
y_test = keras.utils.to_categorical(y_test, num_classes=num_classes)
num_classes = y_test.shape[1]
```

```python
len(X_train)
```

```python
len(np.unique(y_test))
```

```python
# create neural network model
# -----------
num_classes = y_test.shape[1]
model = Sequential()
# Combine the features
num_features = len(X_train.columns)

model = Sequential()
model.add(Dense(64, input_dim=num_features, kernel_initializer='normal', activation='relu'))
model.add(Dense(32, kernel_initializer='normal', activation='relu'))
model.add(Dense(num_classes, kernel_initializer='normal', activation='sigmoid'))  # Use sigmoid for binary classification

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=10, batch_size=32, verbose=1)
```

### Neural Network on Principal Components + Cluster Membership

## Discussion Questions

1. In your own words, what is the difference between PCA and clustering?

PCA reduces the number of features in a dataset by capturing the most important ones, preserving the maximum data variance. Clustering groups similar data points based on their inherent characteristics, aiming to find natural patterns or clusters where points in the same cluster are more alike. Although PCA simplifies data by transforming it into a lower-dimensional space, clustering uncovers meaningful relationships within the dataset. While they have distinct purposes, PCA and clustering can work together, like using PCA to prepare data before applying a clustering algorithm.

2. Did you notice any advantages to combining PCA and clustering? If so, what do you think they were? If not, why do you think you didn't see any gains from this combination?

Combining PCA and clustering didn't consistently outperform individual approaches across different metrics. In the PCA-only neural network, it showed good performance in minimizing loss and mean absolute error, indicating effective dimensionality reduction. On the other hand, the clustering-only neural network exhibited competitive accuracy. However, when combining PCA and clustering, while accuracy remained high, there was a slight increase in loss compared to PCA alone. This suggests that the clustering step may introduce some noise or complexity, impacting the model's ability to generalize. The optimal choice between PCA, clustering, or their combination depends on the specific goals and characteristics of the dataset, and fine-tuning parameters could uncover potential improvements.

3. How can unsupervised techniques help with downstream supervised learning tasks when working with "big data?"

Unsupervised techniques significantly enhance downstream supervised learning with big data by reducing dimensionality, aiding in efficient feature engineering, detecting anomalies, and supporting data preprocessing. Methods like PCA and clustering algorithms optimize model training times, handle outliers, and offer valuable insights into data structure. Combining unsupervised and supervised learning in the NHANES dataset allows clustering individuals based on health characteristics, creating distinct profiles. A subsequent predictive model, trained on these clusters, classifies new participants into health categories, aiding in targeted interventions and personalized recommendations, thus enhancing healthcare strategies.
